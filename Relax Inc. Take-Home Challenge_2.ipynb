```python
df1 = pd.read_csv('relax_challenge/takehome_user_engagement.csv',encoding = "ISO-8859-1")
df1.time_stamp = pd.to_datetime(df1.time_stamp)
df1.index=df1.time_stamp
df1.drop(labels='time_stamp',axis=1,inplace=True)
```

```python
df1
```

```python
df_sum = df1.groupby([pd.Grouper(freq='W'),'user_id']).sum()
```

```python
df_sum
```

Now to find out a user_id who has logged into the product on three separate
days in at least one sevenday
period , identify which factors predict future user adoption



```python
df_adopted = df_sum[df_sum.visited>=3].unstack(level=1).melt()
adopted_users = pd.DataFrame(df_adopted.user_id.unique(),index=range(df_adopted.user_id.unique().shape[0]),columns=['user_id'])
```

```python
df_adopted
```

```python
adopted_users
```

```python
df_join = df.merge(adopted_users,how='inner',left_on='object_id',right_on='user_id')
df_join.shape
```
