```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns',122)
pd.set_option('display.max_rows',122)
```

### Percent of null value in each column

```python
df.isnull().sum()*100/len(df)
```

### Remove columns with null values greater than 50%

```python
df=df[df.columns[df.isnull().sum()*100/len(df) < 50]]
df.isnull().sum()*100/len(df)
```

```python
df.dtypes
```

### Determine outliers for various columns

```python
fig=plt.figure(figsize=(20,10))
ax = fig.add_subplot(231)
ax.boxplot(df['CNT_CHILDREN'])
ax.set_xlabel('Children')
ax = fig.add_subplot(232)
ax.boxplot(df['AMT_INCOME_TOTAL'])
ax.set_xlabel('Income')
ax = fig.add_subplot(233)
ax.boxplot(df['AMT_CREDIT'])
ax.set_xlabel('Credit')
ax = fig.add_subplot(234)
ax.boxplot(df['REGION_POPULATION_RELATIVE'])
ax.set_xlabel('Region Population')
ax = fig.add_subplot(235)
ax.boxplot(df['DAYS_EMPLOYED'])
ax.set_xlabel('Employement Days')
plt.show()
```

## Outliers:
     - Number of Children: The highest outlier is close to 18, which could be a possible entry error.
     - Income: There is a high outlier close to 12*1e8, rest of the outlier seem to be normal
     - Credit: While the mean is at ~500K, the outliers lie between ~1700K and 4000K
     - Region Population: There is just one outlier
     - Employement days: One high outlier at 350000 which is a possible entry error

### Bining of Variables

```python
df['AMT_INCOME_TOTAL'].describe()
```

```python
df['DAYS_LAST_PHONE_CHANGE'].describe()
```

```python
cut_labels_salary = ['Below Avg', 'Avg', 'High']
cut_bins = [0, 1.687979e+05, 2.025000e+05,  1.170000e+08]
df['Salary Bracket'] = pd.cut(df['AMT_INCOME_TOTAL'], bins=cut_bins, labels=cut_labels_salary)

cut_labels_phone = ['Very Old','Old','New','Recent']
cut_bins = [-4292.000000, -1570.000000, -962.858788,-274.000000,0.000000]
df['Phone Age'] = pd.cut(df['DAYS_LAST_PHONE_CHANGE'], bins=cut_bins, labels=cut_labels_phone)
```

### Calculate imbalance and split into two-halves

```python
df_1=df[df['TARGET']==1]
df_0=df[df['TARGET']==0]
print(df_1.shape[0]/df.shape[0]*100)
print(df_0.shape[0]/df.shape[0]*100)
```

## Defining functions for various plots

```python
def cntplt(col,figx=10,figy=5):
    plt.figure(figsize=(figx,figy))
    plt.subplot(1,2,1)
    plt.title('Defaulters')
    sns.countplot(x=col,data=df_1)
    plt.subplot(1,2,2)
    plt.title('Non-Defaulters')
    sns.countplot(x=col,data=df_0)
    plt.show()
def sctplt(col1,col2,figx=10,figy=5):
    plt.figure(figsize=(figx,figy))
    plt.subplot(1,2,1)
    plt.title('Defaulters')
    sns.scatterplot(x=col1,y=col2,data=df_1)
    plt.subplot(1,2,2)
    plt.title('Non-Defaulters')
    sns.scatterplot(x=col1,y=col2,data=df_0)
    plt.show()
def bxplt(col1,col2,figx=10,figy=5):
    plt.figure(figsize=(figx,figy))
    plt.subplot(1,2,1)
    plt.title("Defaulters")
    sns.boxplot(x = col1, y = col2, data = df_1)
    plt.subplot(1,2,2)
    plt.title("Non-Defaulters")
    sns.boxplot(x=col1,y=col2, data = df_0)
    plt.show()
def catcat(col1,col2,figx=10,figy=5):
    plt.figure(figsize = (figx,figy))
    plt.subplot(1,2,1)
    plt.title("Defaulters")
    sns.countplot(x=col1,hue = col2,  data = df_1, palette = 'Set1')
    plt.subplot(1,2,2)
    plt.title("Non-Defaulters")
    sns.countplot(x = col1,hue = col2, data = df_0)
    plt.show()
```
